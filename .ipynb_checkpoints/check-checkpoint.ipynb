```python
import pandas as pd
import numpy as np
```

```python
co2_df = pd.read_csv("./processed_data/co2_emissions.csv")
temp_df = pd.read_csv("./processed_data/Temperature_of_countries.csv")
```

```python
temp_df
```

```python
co2_df
```

```python
t = temp_df.groupby('Year')['Temperature'].mean()
```

```python
c = co2_df.groupby('Year')['Annual CO2 emissions (per capita)'].mean()
```

```python
dataset = pd.DataFrame({
    'CO2 Emissions': c,
    'Temperature': t
})
```

```python
dataset
```

```python
methane_df = pd.read_csv("./processed_data/methane_emissions.csv")
methane_df
```

```python
m = methane_df.groupby('Year')['Per-capita methane emissions in CO2 equivalents'].mean()
m.name = 'MH4 Emissions'
```

```python
m
```

```python
dataset['MH4 Emissions'] = m
```

```python
dataset
```

```python
ni_df = pd.read_csv("./processed_data/nitrous_oxide_emission.csv")
ni_df
```

```python
n = ni_df.groupby('Year')['Per-capita nitrous oxide emissions in CO2 equivalents'].mean()
n
```

```python
dataset['N2O Emissions'] = n
```

```python
dataset
```

```python
population_df = pd.read_csv("./processed_data/population.csv")
population_df
```

```python
p = population_df.groupby('Year')['Population - Sex: all - Age: all - Variant: estimates'].mean()
p
```

```python
dataset['Population'] = p
```

```python
dataset = dataset.loc['1961': '2021', :]
```

```python
dataset
```

```python
dataset.to_csv("dataset/timeseries.csv")
```
