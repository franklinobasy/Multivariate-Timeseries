```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from statsmodels.tsa.api import VAR
```

```python
data = pd.read_csv("dataset/timeseries.csv")
data
```

```python
features = ['CO2 Emissions','MH4 Emissions','N2O Emissions']
# features = ['CO2 Emissions','MH4 Emissions','N2O Emissions','Population']
target = ['Temperature']
```

```python
X = data[features].values  # Input features
y = data[target].values  # Target variable
```

```python
lags = [1, 2, 3, 4, 5]
fig, ax = plt.subplots(len(lags), 1, figsize=(10, 20))

for lag in lags:
    X_lagged = np.zeros((len(X) - lag, lag * X.shape[1]))
    y_lagged = np.zeros(len(y) - lag)

    for i in range(lag, len(X)):
        X_lagged[i - lag] = X[i - lag:i].flatten()
        y_lagged[i - lag] = y[i]


    X_train, X_test, y_train, y_test = train_test_split(X_lagged, y_lagged, test_size=0.2, shuffle=False)

    scaler = MinMaxScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)


    var_model = VAR(np.concatenate((X_train_scaled, y_train.reshape(-1, 1)), axis=1))
    var_results = var_model.fit(maxlags=lag)
    
    lagged_endog = np.concatenate((X_test_scaled[-lag:], y_test[-lag:].reshape(-1, 1)), axis=1)
    y_pred = var_results.forecast(y=lagged_endog, steps=len(y_test))
    mse = mean_squared_error(y_test, y_pred[:, -1])
    print("Mean Squared Error:", mse)
    
    ax[lag - 1].plot(data['Year'][-len(y_test):], y_test, label='Actual Temperature', linewidth=2, marker="o")
    ax[lag - 1].plot(data['Year'][-len(y_test):], y_pred[:, -1], label='Predicted Temperature', linewidth=2, marker="x")
    ax[lag - 1].set_xlabel('Sample', fontsize=12)
    ax[lag - 1].set_ylabel('Temperature', fontsize=12)
    ax[lag - 1].set_title(f"Lag {lag} ---> MSE: {mse:.5f}", fontsize=14, fontweight='bold')
    ax[lag - 1].legend(loc='upper right', fontsize=10)
    ax[lag - 1].tick_params(axis='both', which='major', labelsize=10)
    ax[lag - 1].grid(True, linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
```
